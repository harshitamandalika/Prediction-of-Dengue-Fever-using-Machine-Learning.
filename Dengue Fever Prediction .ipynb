```python
#Loading the dataset
import pandas as pd
import numpy as np
features=pd.read_csv('DengAI/DengAI_Predicting_Disease_Spread_-_Training_Data_Features.csv')
labels=pd.read_csv('DengAI/DengAI_Predicting_Disease_Spread_-_Training_Data_Labels.csv')
data = pd.merge(features,labels,on=['city','year','weekofyear'])
print(data.shape)
np.transpose(data.describe())
```

```python
data
```

```python
#Dividing the data based on the 2 cities 'sj'(San Juan) and 'iq'(Iquitos)
data_sj = data[data['city'] == 'sj']
data_iq = data[data['city'] == 'iq']
```

```python
data_sj
```

```python
data_iq
```

```python
#Identifying the number of null values in each feature of 'iq' data
data_iq.isnull().sum()
```

```python
#Filling the null values in both 'sj' and 'iq' data with their respective mean values 
df_sj=data_sj.fillna(data.mean())
df_iq=data_iq.fillna(data.mean())
```

```python
#Changing temperature scale from Kelvin to Celsius
df_sj['reanalysis_air_temp_k'] = df_sj.reanalysis_air_temp_k -273.15
df_sj['reanalysis_avg_temp_k'] = df_sj.reanalysis_avg_temp_k-273.15
df_sj['reanalysis_dew_point_temp_k'] = df_sj.reanalysis_dew_point_temp_k-273.15
df_sj['reanalysis_max_air_temp_k'] = df_sj.reanalysis_max_air_temp_k-273.15
df_sj['reanalysis_min_air_temp_k'] = df_sj.reanalysis_min_air_temp_k-273.15
```

```python
#Changing temperature scale from Kelvin to Celsius
df_iq['reanalysis_air_temp_k'] = df_iq.reanalysis_air_temp_k -273.15
df_iq['reanalysis_avg_temp_k'] = df_iq.reanalysis_avg_temp_k-273.15
df_iq['reanalysis_dew_point_temp_k'] = df_iq.reanalysis_dew_point_temp_k-273.15
df_iq['reanalysis_max_air_temp_k'] = df_iq.reanalysis_max_air_temp_k-273.15
df_iq['reanalysis_min_air_temp_k'] = df_iq.reanalysis_min_air_temp_k-273.15
```

```python
#Dropping unnecessary features
df_sj=df_sj.drop(['city', 'year' ,'week_start_date'], axis = 1)
df_iq=df_iq.drop(['city', 'year' ,'week_start_date'], axis = 1)
```

```python
#Rounding the data to 2 decimal values
df_sj=df_sj.round(2)
df_iq=df_iq.round(2)
```

```python
#Removing Outliers from 'sj' data using Z-Score
df_zscore_sj = (df_sj - df_sj.mean())/df_sj.std()
print(np.where(df_zscore_sj > 3))
df_sj_z = df_sj[(df_zscore_sj < 3).all(axis=1)]
df_sj_z
```

```python
#Removing Outliers from 'iq' data using Z-Score
df_zscore_iq = (df_iq - df_iq.mean())/df_iq.std()
print(np.where(df_zscore_iq > 3))
df_iq_z = df_iq[(df_zscore_iq < 3).all(axis=1)]
df_iq_z
```
